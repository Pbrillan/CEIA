```python
img1 = cv.cvtColor(cv.imread('CoordCrom_1.png'),cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(cv.imread('CoordCrom_2.png'),cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(cv.imread('CoordCrom_3.png'),cv.COLOR_BGR2RGB)
img4 = img1.copy()
img5 = img2.copy()
img6 = img3.copy()
```

```python
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img1)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img2)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img3)
plt.show()
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img4)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img5)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img6)
plt.show()
```

```python
for i in range(img4.shape[0]):
    for j in range(img4.shape[1]):
        a =(img1[i,j,0]+img1[i,j,1]+img1[i,j,2])
        if a > 0 & a < 765 :
            img4[i,j,0] = 255 *  img1[i,j,0]/(img1[i,j,0]+img1[i,j,1]+img1[i,j,2])
            img4[i,j,1] = 255 *  img1[i,j,1]/(img1[i,j,0]+img1[i,j,1]+img1[i,j,2])
            img4[i,j,2] = 255 *  img1[i,j,2]/(img1[i,j,0]+img1[i,j,1]+img1[i,j,2])
        else :
            img4[i,j,0] = 255 
            img4[i,j,1] = 255
            img4[i,j,2] = 255
            
        a = (img2[i,j,0]+img2[i,j,1]+img2[i,j,2])
        if a > 0 & a < 765:        
            img5[i,j,0] = 255 * img2[i,j,0]/(img2[i,j,0]+img2[i,j,1]+img2[i,j,2])
            img5[i,j,1] = 255 * img2[i,j,1]/(img2[i,j,0]+img2[i,j,1]+img2[i,j,2])
            img5[i,j,2] = 255 * img2[i,j,2]/(img2[i,j,0]+img2[i,j,1]+img2[i,j,2])
        else :
            img5[i,j,0] = 255
            img5[i,j,1] = 255
            img5[i,j,2] = 255
            
        a = (img3[i,j,0]+img3[i,j,1]+img3[i,j,2])
        if a > 0 & a < 765:    
            img6[i,j,0] = 255 * img3[i,j,0]/(img3[i,j,0]+img3[i,j,1]+img3[i,j,2])
            img6[i,j,1] = 255 * img3[i,j,1]/(img3[i,j,0]+img3[i,j,1]+img3[i,j,2])
            img6[i,j,2] = 255 * img3[i,j,2]/(img3[i,j,0]+img3[i,j,1]+img3[i,j,2])
        else :
            img6[i,j,0] = 255
            img6[i,j,1] = 255
            img6[i,j,2] = 255
         
```

```python
maxr=np.amax(img10[:,:,0])
maxg=np.amax(img10[:,:,1])
maxb=np.amax(img10[:,:,2])


sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img10[:,:,0]*sr
G=img10[:,:,1]*sg
B=img10[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img13=cv.merge([R,G,B])

maxr=np.amax(img11[:,:,0])
maxg=np.amax(img11[:,:,1])
maxb=np.amax(img11[:,:,2])
sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img11[:,:,0]*sr
G=img11[:,:,1]*sg
B=img11[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img14=cv.merge([R,G,B])

maxr=np.amax(img12[:,:,0])
maxg=np.amax(img12[:,:,1])
maxb=np.amax(img12[:,:,2])
sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img12[:,:,0]*sr
G=img12[:,:,1]*sg
B=img12[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img15=cv.merge([R,G,B])
```
