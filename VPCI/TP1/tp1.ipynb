```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img10 = cv.cvtColor(cv.imread('WP_B.png'),cv.COLOR_BGR2RGB)
img11 = cv.cvtColor(cv.imread('WP_O.png'),cv.COLOR_BGR2RGB)
img12 = cv.cvtColor(cv.imread('WP_R.png'),cv.COLOR_BGR2RGB)
```

```python
maxr=np.amax(img10[:,:,0])
maxg=np.amax(img10[:,:,1])
maxb=np.amax(img10[:,:,2])


sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img10[:,:,0]*sr
G=img10[:,:,1]*sg
B=img10[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img13=cv.merge([R,G,B])

maxr=np.amax(img11[:,:,0])
maxg=np.amax(img11[:,:,1])
maxb=np.amax(img11[:,:,2])
sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img11[:,:,0]*sr
G=img11[:,:,1]*sg
B=img11[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img14=cv.merge([R,G,B])

maxr=np.amax(img12[:,:,0])
maxg=np.amax(img12[:,:,1])
maxb=np.amax(img12[:,:,2])
sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print(maxr,maxg,maxb)
print(sr,sg,sb)
R=img12[:,:,0]*sr
G=img12[:,:,1]*sg
B=img12[:,:,2]*sb
R=R.astype('uint8')
G=G.astype('uint8')
B=B.astype('uint8')
img15=cv.merge([R,G,B])
```

```python
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img13)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img14)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img15)
plt.show()
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img10)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img11)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img12)
plt.show()
```
