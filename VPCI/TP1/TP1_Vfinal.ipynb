```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

COORDENADAS CROMATICAS

```python
img1 = cv.cvtColor(cv.imread('CoordCrom_1.png'),cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(cv.imread('CoordCrom_2.png'),cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(cv.imread('CoordCrom_3.png'),cv.COLOR_BGR2RGB)
img4 = img1.copy()
img5 = img2.copy()
img6 = img3.copy()
```

```python
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img1)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img2)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img3)
plt.show()
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img4)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img5)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img6)
plt.show()
```

```python
##Metodo Compacto con correcciones

B,G,R = cv.split(img4)
image_norm = np.sum(img4,axis=2)
image_norm[image_norm == 0]= 1
B = (B/image_norm)*255
B=B.astype('uint8')
R = (R/image_norm)*255
R=R.astype('uint8')
G = (G/image_norm)*255
G=G.astype('uint8')
img4=cv.merge([B,G,R])

B,G,R = cv.split(img5)
image_norm = np.sum(img5,axis=2)
image_norm[image_norm == 0]= 1
B = (B/image_norm)*255
B=B.astype('uint8')
R = (R/image_norm)*255
R=R.astype('uint8')
G = (G/image_norm)*255
G=G.astype('uint8')
img5=cv.merge([B,G,R])

B,G,R = cv.split(img6)
image_norm = np.sum(img6,axis=2)
image_norm[image_norm == 0]= 1
B = (B/image_norm)*255
B=B.astype('uint8')
R = (R/image_norm)*255
R=R.astype('uint8')
G = (G/image_norm)*255
G=G.astype('uint8')
img6=cv.merge([B,G,R])
```

```python
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img4)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img5)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img6)
plt.show()
fig = plt.figure()
ax3 = plt.subplot(131)
ax1 = plt.imshow(img1)
ax2 = plt.subplot(132)
ax2 = plt.imshow(img2)
ax1 = plt.subplot(133)
ax3 = plt.imshow(img3)
plt.show()
```
