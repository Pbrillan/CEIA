```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
imagen = cv.imread('eyes.jpg')
img_RGB = cv.cvtColor(imagen, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(imagen, cv.COLOR_BGR2GRAY)
```

```python
img_gray_soft = cv.medianBlur(img_gray,5)

fig, ax = plt.subplots(1,3,figsize=(18,6)) 

ax[0].set_title("Imagen Original")
ax[0].imshow(img_RGB)
ax[0].axis('off')
ax[1].set_title("Imagen en Escala de Grises")
ax[1].imshow(img_gray,cmap='gray')
ax[1].axis('off')
ax[2].set_title("Imagen en Escala de Grises Blur")
ax[2].imshow(img_gray_soft,cmap='gray')
ax[2].axis('off')
plt.show()
```

```python
img_iris = img_RGB.copy()
circulos_iris = cv.HoughCircles(img_gray_soft, 
                                    cv.HOUGH_GRADIENT, 
                                    1,                  
                                    90,                
                                    param1=160,         
                                    param2=15,          
                                    minRadius=28,       
                                    maxRadius=31)   
```

```python
#print(circulos_iris)
```

```python
for i in circulos_iris[0,:]:
    # draw the outer circle
    cv.circle(img_iris,(i[0],i[1]),i[2],(0,255,255),2)
    cv.circle
plt.figure(figsize=(10, 12))
plt.imshow(img_iris,cmap='gray')
plt.show()
```

```python
cicles_oy = np.argsort(circulos_iris[0,:,1])
circulos_pupy = circulos_iris[:,cicles_oy,:] 
img_iris_dist = img_iris.copy()
distancia=np.array(range(int(cicles_oy.size/2)))
for i in range(int(cicles_oy.size/2)):
    eye1 = circulos_pupy[:,i*2,:]
    eye2 = circulos_pupy[:,i*2+1,:]
    if eye1[0,0] > eye2[0,0]:
        distancia[i]=eye1[0,0] - eye2[0,0]
    else:
        distancia[i]=eye2[0,0] - eye1[0,0]

    cv.putText(img_iris_dist,str(distancia[i]),(265 ,eye1[0,1]),cv.FONT_HERSHEY_SIMPLEX,1,(0,255,255),2,cv.LINE_AA)
        
print(distancia)
plt.figure(figsize=(10, 12))
plt.imshow(img_iris_dist,cmap='gray')
plt.show()
```

```python
img_pupilas = img_RGB.copy()
circulos_pupilas = cv.HoughCircles(img_gray_soft, 
                                    cv.HOUGH_GRADIENT, 
                                    1,                  
                                    90,                
                                    param1=160,         
                                    param2=15,          
                                    minRadius=10,       
                                    maxRadius=15)       
```

```python
#print(circulos_pupilas)
```
